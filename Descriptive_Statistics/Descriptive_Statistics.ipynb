```python
arr=[20,56,19,63,79,99,37,80,999,419,88,34]
ar=pd.DataFrame(arr)
ar
```

```python
length=len(arr)
```

```python
length
```

```python
sum = 0
for i in range ( 0, length ):
    sum+=arr[i]
print("Sum: ",sum,"\n")
avg = sum/length
print("Mean: ",avg)
```

```python
num = 0
for i in range ( 0, length ):
    num+=(arr[i]-avg)**2
var=num/(length-1)
print("Varience: ",var,"\n")
std=var**0.5
print("std: ",std)
```
